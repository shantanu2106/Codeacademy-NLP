```python
dataset = pd.read_csv('/Users/shantanusharma/Desktop/Codeacademy projects/Codeacademy-NLP/US Medical Insurance Project/insurance.csv')
dataset.head()
```

Loading the columns to list form

```python
age = list(dataset['age'])
sex = list(dataset['sex'])
bmi = list(dataset['sex'])
children = list(dataset['children'])
smoker = list(dataset['children'])
region = list(dataset['region'])
charges = list(dataset['charges'])
columns = list(dataset.columns)
```

# Defining the Scope of the project
* find average age of the patients
* return the number of males vs. females counted in the dataset
* find geographical location of the patients
* return the average yearly medical charges of the patients
* creating a dictionary that contains all patient information

```python
# Finding average age of the patients
average_age = sum(age)/len(age)
print(average_age)
# Number of males vs females in dataset
male_count = sex.count('male')
female_count = len(sex) - male_count
print(male_count,female_count)
```

```python
# Finding Geographical Location of the patients - Using Dictionary
regions = list(set(region))
area_dict = {}
for area in regions:
    area_count = region.count(str(area))
    area_dict[area] = area_count
print(area_dict)
```

```python
# Average Yearly Medical Charges
average_charges = sum(charges)/len(charges)
print(average_charges)
```

```python
# Creating a dictionary that contains all patient information
patient_info_list = []
patient_info_dict = {}
for element in range(len(age)):
    patient_info = {
        'Age' : age[element],
        'Sex' : sex[element],
        'BMI' : bmi[element],
        'Children' : children[element],
        'Smoker' : smoker[element],
        'Charges' : charges[element]}
    patient_info_list.append(patient_info)
    patient_name = 'Patient'+str(element)
    patient_info_dict[patient_name] = patient_info_list[element]
print(patient_info_dict)
```
